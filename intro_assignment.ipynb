```python

""" Function to permute plaintext based on p-table elements
    here each element in the plaintext is transpositioned based on their index values in the p-table"""
def p_table(plaint_txt):
    print("Enter P-Table elements(1-16): ")
    p_table  = []
    for i in range(0,16):
        x = int(input())
        p_table.append(x)

    new_txt = ""
    for i in p_table:
        new_txt += plain_txt[i - 1]
    return new_txt,p_table

#This function is used to split plaintext into two halves
def split(new_txt):
    L0 = new_txt[0:8]
    R0 = new_txt[8:16]
    return L0,R0

#This function is used to expand 8-bit input into 12-bit output
def expnd(R0):
    expand = []
    print("Enter expansion table elements(1-8): ")
    for i in range(0,12):
        x = int(input())
        expand.append(x)

    expanded_txt = ""
    for j in expand:
        expanded_txt += R0[j - 1]

    return expanded_txt

#this function is used to perform XOR operation on two values
def XOR(a, b):
    xor_txt = ""
    for x,y in zip(a, b):
        sum = (int(x) + int(y)) % 2
        if sum == 1:
            xor_txt += str(1)
        else:
            xor_txt += str(0)
    return xor_txt

"""S-box is used to used to get new 8-bit value from the 12-bit value obtained by XOR between
 expanded 12-bits and key."""
def s_box(xor_txt):
    s_box = [[6,4,15,13,1,10,5,3,11,7,12,0,9,14,2,8],
             [3,2,0,13,8,6,1,15,12,4,5,14,10,7,11,9],
             [1,9,7,3,11,5,15,10,14,12,8,4,13,6,2,0],
             [11,0,10,6,1,12,15,2,7,13,3,9,14,4,8,5]]

    B1 = xor_txt[0:6]
    B2 = xor_txt[6:12]

    row1 = B1[0] + B1[5]
    col1 = B1[1] + B1[2] + B1[3] + B1[4]  #col1 in binary form

    row1_sum = 0
    index = 0
    for k in range(2,0,-1):
        row1_sum += int(row1[index])*(2**(k-1))    #convertion of row_sum into binary value
        index+=1

    col1_sum = 0
    index = 0
    for k in range(4,0,-1):
        col1_sum += int(col1[index])*(2**(k-1))
        index+=1

    print("Row1: ",row1_sum," Col1: ",col1_sum)

    row2 = B2[0] + B2[5]
    col2 = B2[1] + B2[2] + B2[3] + B2[4]

    row2_sum = 0
    index = 0
    for k in range(2,0,-1):
        row2_sum += int(row2[index])*(2**(k-1))
        index+=1

    col2_sum = 0
    index = 0
    for k in range(4,0,-1):
        col2_sum += int(col2[index])*(2**(k-1))
        index+=1

    print("Row2: ",row2_sum," Col2: ",col2_sum)
    print("\n")

    B1_value = s_box[row1_sum][col1_sum]
    print("B1 segment value: ",B1_value)

    B2_value = s_box[row2_sum][col2_sum]
    print("B2 segment value: ",B2_value)
    print("\n")

    return B1_value,B2_value

#function to convert decimal to binary
def convert_to_binary(B1_value,B2_value):
    B1_bin_temp = []
    while B1_value > 0:
        rem = B1_value % 2
        B1_bin_temp.append(str(rem))
        B1_value //= 2

    B1_bin = ''.join(B1_bin_temp[::-1])
    print("Binary value of B1: ",B1_bin)

    rem = 0
    B2_bin_temp = []
    while B2_value > 0:
        rem = B2_value % 2
        B2_bin_temp.append(str(rem))
        B2_value //= 2

    B2_bin = ''.join(B2_bin_temp[::-1])
    print("Binary value of B2: ",B2_bin)

    new_8bit = str(B1_bin + B2_bin.zfill(4))
    print("8-bit after s-box: ",new_8bit)
    print("\n")

    return new_8bit

#this function takes 8-bit output from s-box as input and permutes it based on its index value in the p-box
def p_box(new_8bit):
    pbox = []
    print("Enter P-Box elements(1-8): ")
    for i in range(0,8):
        x = int(input())
        pbox.append(x)
    pbox_8bit = ""
    for i in pbox:
        pbox_8bit += new_8bit[i - 1]

    print("After p-box: ",pbox_8bit)
    print("\n")
    return pbox_8bit

#this function is used to inverse the permutation table
def p_inverse(p_table, L1R1):
    pinverse = []
    for i in range(1,17):
        j = i
        for j in p_table:
            if i == p_table[j-1]:
                pinverse.append(j)
                break

    cipher_txt = ""
    for k in pinverse:

        cipher_txt += L1R1[int(k) - 1]

    print("Cipher text: ",cipher_txt)

    return cipher_txt


#main fn
plain_txt = input("Enter the plain text(16-bit): ")
key = input("Enter the key(12-bit): ")

new_txt,ptable = p_table(plain_txt)
print("After p-table: ",new_txt)

L0,R0 = split(new_txt)
print("L0: ",L0)
print("R0: ",R0)

expanded_txt = expnd(R0)
print("After expansion to 12-bits: ",expanded_txt)

xor_txt = XOR(expanded_txt, key)
print("After XOR op with key: ",xor_txt)

B1_value,B2_value = s_box(xor_txt)

new_8bit = convert_to_binary(B1_value,B2_value)

pbox_8bit = p_box(new_8bit)

xor_txt = XOR(pbox_8bit,L0)

L1 = R0
R1 = xor_txt
print("L1: ",L1)
print("R1: ",R1)

L1R1 = L1 + R1
print("L1R1:",L1R1)
print("\n")

cipher_txt = p_inverse(ptable, L1R1)
```

```python
#This function slices the plain text into blocks of size:4
def slice(plain_txt):
        substr = []
        for i in range(0, len(plain_txt), 4):
            substr.append(plain_txt[i:i+4])
        return substr

#converts decimal to binary
def convert_to_binary(ctr):
    bin_str = format(ctr, 'b')
    return bin_str.zfill(4)

#performs XOR betweeen message and counter value
def XOR(a, b):
    xor_txt = ""
    for x,y in zip(a, b):
        sum = (int(x) + int(y)) % 2
        if sum == 1:
            xor_txt += str(1)
        else:
            xor_txt += str(0)
    return xor_txt

#performs encryption based on the ondex value specifies in key
def enc(x, new_txt, key):
    for i in key:
        new_txt += x[i - 1]
    return new_txt

def encryption():
    plain_txt = input("Enter plain text(block size:4): ")
    block = slice(plain_txt)
    ctr = int(input("Enter counter value: "))
    ctr_bin = convert_to_binary(ctr)
    key = [3,2,1,4]

    cipher = ""
    temp = []
    for i in block:
        xor_result = XOR(i, ctr_bin)
        temp.append(xor_result)
        print(ctr_bin)
        ctr += 1                               #after each iteration, counter value is incremented by 1
        ctr_bin = convert_to_binary(ctr)
    print(temp)

    new_txt = ""
    for i in temp:
        result = enc(i, new_txt, key)
        cipher += result
    print("Cipher Text: ",cipher)

def decryption():
    cipher_txt = input("Enter cipher text(block size:4): ")
    block = slice(cipher_txt)
    ctr = int(input("Enter counter value: "))
    ctr_bin = convert_to_binary(ctr)
    key = [3,2,1,4]

    plain = ""
    new_txt = ""
    temp = ""
    for i in block:
        result = enc(i, new_txt, key)
        plain += result


    for i in range(0, len(plain), 4):
        block_plain = plain[i:i+4]
        xor_result = XOR(block_plain, ctr_bin)
        temp += xor_result
        ctr += 1
        ctr_bin = convert_to_binary(ctr)
    print("Plain text: ",temp)

opt = int(input(("""
             1. Encryption
             2. Decryption
             3. Exit
             Enter your option: """)))
if opt == 1:
    encryption()
elif opt == 2:
    decryption()
elif opt == 3:
    exit()
else:
    print("Invalid option!!!")
```
